```python
from ultralytics import YOLO

# Load YOLOv8 model (v8n = nano, v8s = small, v8m = medium)
model = YOLO('yolov8n.pt')  # Starting with a lightweight model

# Set up training parameters
model.train(
    data='/content/data/data.yaml',      # Path to YAML file with dataset information
    epochs=50,                    # Number of epochs
    batch=16,                     # Batch size
    imgsz=640,                    # Image size
    augment=True,                 # Enable data augmentation
    project='yolo_training',      # Project name
    name='yolov8_aquarium'        # Experiment name
)
```

# Answers to Questions
# Using augment=True to apply random transformations (flips, rotations, etc.) to make the model robust.


# Starting with 50 epochs. Will increase if performance plateaus early or requires improvement.


# Batch size 16 and image size 640 selected for balanced training efficiency and memory use.


# Analyzing mAP@50, Precision, Recall to evaluate detection accuracy and model reliability. After training, a higher mAP@50-95 score shows that the model maintains good detection accuracy across stricter IoU thresholds.


#Concrete result: For this dataset, a high precision score (e.g., >0.85) would mean that when the model predicts an object, it’s often correct.

```python
# Run validation to analyze results
metrics = model.val()
print(metrics)
```
